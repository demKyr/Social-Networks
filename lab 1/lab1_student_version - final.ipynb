## Α. Κατασκευή και οπτικοποίηση σύνθετων τύπων δικτύου

### Τοπολογία REG για βαθμό πλέγματος $d = 4$

```python
n = 41 # μέγεθος δικτύου: αντικαταστήστε το x με το τελευταίο ψηφίο του αριθμού μητρώου σας.
```

```python
random.seed(seedno)
d=4

# για την κατασκευή του REG χρησιμοποιήστε την watts_strogatz_graph από το networkx. 
# Επιλέξτε προσεκτικά την τιμή της παραμέτρου p. 
REG = nx.watts_strogatz_graph(n, d, 0, seedno) 

#οπτικοποίηση: χρησιμοποιήστε την draw_circular από το networkx
nx.draw_circular(REG,with_labels=True)
```

Παρατήρηση: Κάθε κόμβος ενώνεται εκτός από τους δύο διπλανούς του και με τους δύο επόμενους αν και δεν φαίνεται ξεκάθαρα στο πιο πάνω διάγραμμα

### Τοπολογία RG (ER) για αριθμό συνδέσεων $M = 150$

```python
random.seed(seedno)

M = 150

# για την κατασκευή του RGER χρησιμοποιήστε την gnm_random_graph από το networkx
RGER = nx.gnm_random_graph(n,M,seedno) 

# οπτικοποίηση: χρησιμοποιήστε την draw_circular από το networkx
nx.draw_circular(RGER,with_labels=True)
```

### Τοπολογία RGG  με ακτίνα σύνδεσης $R=0.35$

```python
random.seed(seedno)

R=0.35

# για την κατασκευή του RGG χρησιμοποιήστε την random_geometric_graph από το networkx
RGG = nx.random_geometric_graph(n, R, seed=seedno)

#οπτικοποίηση: https://networkx.org/documentation/stable/auto_examples/drawing/plot_random_geometric_graph.html
pos = nx.get_node_attributes(RGG, "pos")

# find node near center (0.5,0.5)
dmin = 1
ncenter = 0
for p in pos:
    x, y = pos[p]
    d = (x - 0.5) ** 2 + (y - 0.5) ** 2
    if d < dmin:
        ncenter = p
        dmin = d

# color by path length from node near center
p = dict(nx.single_source_shortest_path_length(RGG, ncenter))

plt.figure(figsize=(8, 8))
nx.draw_networkx_edges(RGG, pos, alpha=0.4)
nx.draw_networkx_nodes(
    RGG,
    pos,
    nodelist=list(p.keys()),
    node_size=80,
    node_color=list(p.values()),
    cmap=plt.cm.Reds_r,
)

plt.xlim(-0.05, 1.05)
plt.ylim(-0.05, 1.05)
# plt.axis("off")
plt.show()
```

### Τοπολογία SF με παράμετρο επισύναψης $d=4$

```python
random.seed(seedno)

d=4

# για την κατασκευή του SF χρησιμοποιήστε την barabasi_albert_graph από το networkx
SF =  nx.barabasi_albert_graph(n, d, seedno)

#οπτικοποίηση: χρησιμοποιήστε την draw_circular από το networkx
nx.draw_circular(SF,with_labels=True)
```

### Τοπολογία SW με βαθμό αρχικού πλέγματος $d=4$ και πιθανότητα ανασύνδεσης $p=0.25$

```python
random.seed(seedno)

d=4
p=0.25

# για την κατασκευή του SW χρησιμοποιήστε την watts_strogatz_graph από το networkx. 
SW = nx.watts_strogatz_graph(n=n,k=d,p=p,seed=seedno) 

#οπτικοποίηση: χρησιμοποιήστε την draw_circular από το networkx
nx.draw_circular(SW,with_labels=True)
```

## B. Μετρικές δικτύου

### B1. Συντελεστής ομαδοποίησης (Clustering coefficient)

#### Για κάθε μια από τις τοπολογίες REG, RG(ER), RGG, SF(BA) και SW(WS):
#### Β1.1 Να υπολογιστεί ο συντελεστής ομαδοποίησης κάθε κόμβου και να αναπαρασταθεί η κατανομή του.
#### Β1.2  Να υπολογιστεί ο μέσος συντελεστής ομαδοποίησης και η διασπορά του.
#### Β.1.3 Να βρεθούν οι κόμβοι με το μικρότερο και το μεγαλύτερο συντελεστή ομαδοποίησης. 
#### Να σχολιαστούν τα αποτελέσματα.

```python
def cc(G):
    return nx.clustering(G)

def clustering_coef_info(G_clust_coef,name):
    print(name)
    G_clust_coef_val = np.array(list(G_clust_coef.values())) 
    # B1.1 Histogram of clustering coefficients
    plt.figure()
    plt.title("Histogram of Clustering coefficients for " + name)
    plt.ylabel("# of nodes")
    plt.xlabel("Clustering coefficient value")
    plt.hist(G_clust_coef_val, range = (0,1), bins = 100)
    # B1.2 Calculate mean value and variance of clustering coefficients
    print("Mean value of clustering coefficients:", np.mean(G_clust_coef_val))
    print("Variance of clustering coefficients:",np.var(G_clust_coef_val))
    # B1.3 Calculate max and min value of clustering coefficients
    max_val = np.amax(G_clust_coef_val)
    print("Node(s) with maximum clustering coefficient value = ",max_val , {k for k,v in G_clust_coef.items() if v==max_val})
    min_val = np.amin(G_clust_coef_val)
    print("Node(s) with minimum clustering coefficient value = ",min_val , {k for k,v in G_clust_coef.items() if v==min_val})
```

```python
clustering_coef_info(cc(REG),"REG")
clustering_coef_info(cc(RGER),"RGER")
clustering_coef_info(cc(RGG),"RGG")
clustering_coef_info(cc(SF),"SF")
clustering_coef_info(cc(SW),"SW")
```

Παρατηρήσεις: Ο συντελεστής ομαδοποίησης δίνει μια εικόνα για το πόσο ισχυρά είναι συνδεδεμένη η γειτονιά ενός κόμβου
* Για τον REG γράφο όλοι οι κόμβοι έχουν ίδιο συντελεστή ομαδοποίησης κάτι που είναι αναμενόμενο λόγο ομοιομορφίας. Επίσης, έχουν όλοι οι κόμβοι τιμή 0.5 αφου από τους 4 γείτονες που έχει ο κάθε κόμβος και τις 6 τριπλέτες όπου ανήκει, οι 3 αποτελούν τρίγωνα
* Για τον RG(ER) γράφο οι περισσότεροι κόμβοι έχουν μικρό συντελεστή ομαδοποίησης αφού όπως φαίνεται και στο πιο πάνω σχήμα οι ακμές είναι κατανεμημένες τυχαία και δεν υπάρχουν εμφανείς γειτονιές που να είναι συνδεδεμένες
* Για τον RGG γράφο παρατηρούνται οι μεγαλύτερες τιμές συντελεστών ομαδοποίησης διότι οι κόμβοι είναι συνδεδεμένοι με τους γειτονικούς τους δημιουργόντας ισχυρά συνδεδεμένες γειτονίες. Οι κόμβοι που έχουν μικρότερες τιμές στο πιο πάνω ιστόγραμμα είναι οι κόμβοι στο μεταίχμιο μεταξύ των ισχυρά συνδεδεμένων γειτονιών
* Για τον SF γράφο τα αποτελέσματα μοιάζουν με τον RG(ER) επειδή γενικά δεν υπάρχουν εμφανείς γειτονιές
* Για τον SW γράφο υπάρχουν λίγες δυνατές τιμές συντελεστή ομαδοποίησης γιατί αρχικά ο γράφος ήταν REG και άλλαξαν κάποιες ακμές επομένως διατηρεί σε κάποιο βαθμό τον χαρακτήρα του REG αφού σε πολλές περιπτώσεις οι κόμβοι είναι συνδεδεμένοι με τους γειτονικούς τους

### Β2. Μέσο μήκος μονοπατιού - Διάμετρος - Ακτίνα γράφου

#### Για κάθε μια από τις τοπολογίες REG, RG(ER), RGG, SF(BA) και SW(WS):
#### Β2.1 Nα υπολογιστεί το μέσο μήκος μονοπατιού και η διασπορά του. Προτείνεται η χρήση της συνάρτησης  all_pairs_shortest_path_length του networkx.
#### B2.2 Να υπολογιστεί η διάμετρος (συνάρτηση diameter) και η ακτίνα (συνάρτηση radius).
#### Για τα παραπάνω πραγματοποιήστε συγκρίσεις και εξηγήστε με βάση τα χαρακτηριστικά της κάθε τοπολογίας όπως παρουσιάζονται στη θεωρία.

```python
def compute_path_lengths(G):
    return dict(nx.all_pairs_shortest_path_length(G))

def shortest_path_info(G,paths,name):
    print(name)
    paths_flattened = pd.json_normalize(paths, sep='-').to_dict(orient='records')[0]
    paths_flattened_val = np.array(list(paths_flattened.values())) 
    # B2.1 Calculate mean value and variance of path lengths
    print("Average path length:", np.mean(paths_flattened_val))
    print("Variance of path lengths:",np.var(paths_flattened_val))
    # B2.2 Calculate the diameter and the radius of the graph
    print("Diameter of the graph:", nx.diameter(G))
    print("Radius of the graph:", nx.radius(G))
```

```python
shortest_path_info(REG,compute_path_lengths(REG),"REG")
shortest_path_info(RGER,compute_path_lengths(RGER),"RGER")
shortest_path_info(RGG,compute_path_lengths(RGG),"RGG")
shortest_path_info(SF,compute_path_lengths(SF),"SF")
shortest_path_info(SW,compute_path_lengths(SW),"SW")
```

Παρατηρήσεις: Το μέσο μήκος μονοπατιού δίνει μια εικόνα για τη μέση απόσταση δύο κόμβων ενώ η διασπορά δείχνει πόσο απέχουν περίπου οι τιμές των μονοπατιών από τη μέση τιμή. Η διάμετρος είναι το μέγιστο μήκος μονοπατιού σε ένα γράφο ενώ η ακτίνα είναι το μικρότερο μονοπάτι από μια λίστα που περιέχει τις n τιμές των μέγιστων μονοπατιών από κάθε κόμβο.
* Για τον REG γράφο υπάρχει μέγιστη μέση τιμή μονοπατιού και διασποράς του γιατί υπάρχουν μονοπάτια πολύ μικρά και πολύ μεγάλα. Επίσης έχει και τη μέγιστη ακτίνα και διάμετρο αφού κάθε κόμβος απέχει 10 βήματα από τους 4 πιο μακρινούς του γείτονες
* Για τον RG(ER) γράφο η μέση τιμή μονοπατιού είναι μικρή αφού λόγο τυχαιότητας των ακμών υπάρχουν κόμβοι μέσω των οποίων γειτονικοί τους κόμβοι ενώνονται απευθείας με μακρινούς. Παρόλα αυτά υπάρχει μεγάλη διασπορά αφού οι τιμές των μονοπατιών είναι γενικά τυχαίες. 
* Για τον RGG γράφο η μέση τιμή μονοπατιού είναι σχετικά μεγάλη σε σχέση με τους πιο τυχαίους γράφους αφού οι κόμβοι είναι συνδεδεμένοι με τους γειτονικούς τους και για να φτάσουν στους μακρινούς πρέπει να διασχίσουν μεγάλη απόσταση. Επίσης έχουν μεγάλη διασπορά σε σχέση με τους άλλους γιατί για τους κόμβους εντος της γειτονιάς τους τα μονοπάτια είναι πολύ μικρά ενώ για κόμβους άλλων γειτονιών είναι μεγάλα. Για τον ίδιο λόγο έχει και σχετικά μεγάλη διάμετρο
* Για τον SF γράφο η μέση τιμή μονοπατιού και οι διασπορά είναι μικρά διότι οι κόβμοι χρησιμοποιούν τους κόμβους με πολλούς γείτονες ώστε να κινηθούν γρήγορα σε οποιοδήποτε άλλο κόμβο. Έχει μικρή ακτίνα γιατί κάποιοι κόμβοι έχουν πολλές συνδέσεις άρα και μικρά μονοπάτια με όλους τους κόμβους.
* Για τον SW γράφο η μέση τιμή μονοπατιού και οι διασπορά είναι σχετικά μεγάλα γιατί πολλοί κόμβοι διατηρούν το χαρακτήρα του REG ενώ άλλοι έχουν μονοπάτια προς πολλούς και μακρινούς κόμβους. Για τον ίδιο λόγο η διάμετρος είναι μεγάλη

#### B.2.3 Για κάθε τοπολογία, ποιος κόμβος συνδέεται με τους περισσότερους κόμβους με μήκος ελάχιστου μονοπατιού ίσο με τη διάμετρο της τοπολογίας?

```python
def find_nodes(G,name):
    print(name)
    paths = compute_path_lengths(G)
    no_of_diameter_paths_per_node = {}
    starting_nodes_with_dest = {}
    # Calculate number of paths with diameter length from every node 
    for i in range(len(paths)):
        no_of_diameter_paths_per_node[i]=sum(value == nx.diameter(G) for value in paths[i].values())
    # Find labels of nodes with most paths with diameter length
    starting_nodes = {k for k,v in no_of_diameter_paths_per_node.items() if v==max(no_of_diameter_paths_per_node.values())}
    # Create set with the starting nodes and the destination nodes
    for node in starting_nodes:
        starting_nodes_with_dest[node] = {k for k,v in paths[node].items() if v==nx.diameter(G)}
    print("Node(s) with most paths with length equal to the diameter and their destinations", starting_nodes_with_dest)
```

```python
find_nodes(REG,"REG")
find_nodes(RGER,"RGER")
find_nodes(RGG,"RGG")
find_nodes(SF,"SF")
find_nodes(SW,"SW")
```

#### Περιφέρεια του γράφου ορίζεται να είναι το σύνολο των κόμβων του που έχουν εκκεντρότητα ίση με τη διάμετρό του. 
#### Κέντρο του γράφου ορίζεται να είναι το σύνολο των κόμβων του που έχουν εκκεντρότητα ίση με την ακτίνα του.  
#### Β.2.4 Για κάθε τοπολογία, να βρεθεί η περιφέρεια και το κέντρο της και να σχολιαστούν τα αποτελέσματα.

```python
def find_periphery(G):
    return {k for k,v in nx.eccentricity(G).items() if v==nx.diameter(G)}
```

```python
def find_center(G):
    return {k for k,v in nx.eccentricity(G).items() if v==nx.radius(G)}
```

```python
def print_periphery_n_center(G,name):
    print(name)
    print("Set of nodes in periphery", find_periphery(G))
    print("Set of nodes in center", find_center(G))

print_periphery_n_center(REG,"REG")
print_periphery_n_center(RGER,"RGER")
print_periphery_n_center(RGG,"RGG")
print_periphery_n_center(SF,"SF")
print_periphery_n_center(SW,"SW")

```

## Γ) Mετρικές κεντρικότητας κόμβων

### Γ1. Κεντρικότητα βαθμού (Degree centrality)
##### Γ1.1 Να αναπαρασταθεί η κατατανομή βαθμών κόμβων για κάθε τοπολογία.
##### Γ1.2 Να υπολογιστεί ο μέσος βαθμός κόμβου καθώς και η διασπορά των βαθμών κόμβου για κάθε τοπολογία.
##### Να σχολιαστούν τα αποτελέσματα.

```python
def find_degree(G,name):
    print(name)
    G_deg_cent = nx.degree_centrality(G)
    G_deg_cent_val = np.array(list(G_deg_cent.values())) 

    G_deg_cent_non_normalized = G_deg_cent.copy()
    G_deg_cent_non_normalized.update((x, int(y * (n-1))) for x, y in G_deg_cent_non_normalized.items())
    G_deg_cent_non_normalized_val = np.array(list(G_deg_cent_non_normalized.values())) 
    
    print("Average degree:", np.mean(G_deg_cent_non_normalized_val))
    print("Variance of degree:",np.var(G_deg_cent_non_normalized_val))
    print("Average Normalized degree:", np.mean(G_deg_cent_val))
    print("Variance of Normalized degree:",np.var(G_deg_cent_val))
    fig = plt.figure(figsize=(14, 6))
    ax1 = fig.add_subplot(121)
    ax2 = fig.add_subplot(122)
    ax1.set_ylabel("# of nodes")
    ax1.set_xlabel("Normalized degree centrality")
    ax2.set_ylabel("# of nodes")
    ax2.set_xlabel("degree centrality")
    ax1.hist(G_deg_cent_val, range = (0,1), bins = 100)
    ax1.title.set_text('Histogram of Normalized degree centrality for '+ name)
    ax2.hist(G_deg_cent_non_normalized_val, range = (0,n), bins = n)
    ax2.title.set_text('Histogram of Non-Normalized degree centrality for '+ name)
    plt.show()
```

Ο βαθμός κεντρικότητας ενός κόμβου ορίζεται σε μη κανονικοποιημένη μορφή ως ο αριθμός των συνδέσεων ενός κόμβου

### Γ2. Κεντρικότητα εγγύτητας (Closeness centrality)
##### Γ2.1 Να αναπαρασταθεί η κατατανομή της εγγύτητας κόμβων για κάθε τοπολογία.
##### Γ2.2 Να υπολογιστεί η μέση εγγύτητα καθώς και η διασπορά της για κάθε τοπολογία.
##### Να σχολιαστούν τα αποτελέσματα.

```python
def find_closeness(G,name):
    print(name)
    G_closeness_cent = nx.closeness_centrality(G)
    G_closeness_cent_val = np.array(list(G_closeness_cent.values())) 

    G_closeness_cent_non_normalized = G_closeness_cent.copy()
    G_closeness_cent_non_normalized.update((x, (y / (n-1))) for x, y in G_closeness_cent_non_normalized.items())
    G_closeness_cent_non_normalized_val = np.array(list(G_closeness_cent_non_normalized.values())) 

    
    print("Average closseness centrality:", np.mean(G_closeness_cent_non_normalized_val))
    print("Variance of closseness centrality:",np.var(G_closeness_cent_non_normalized_val))
    print("Average normalized closseness centrality:", np.mean(G_closeness_cent_val))
    print("Variance of normalized closseness centrality:",np.var(G_closeness_cent_val))
    fig = plt.figure(figsize=(14, 6))
    ax1 = fig.add_subplot(121)
    ax2 = fig.add_subplot(122)
    ax1.set_ylabel("# of nodes")
    ax1.set_xlabel("Normalized closseness centrality")
    ax2.set_ylabel("# of nodes")
    ax2.set_xlabel("closseness centrality")
    ax1.hist(G_closeness_cent_val, range = (0,1), bins = 100)
    ax1.title.set_text('Histogram of Normalized closseness centrality for '+ name)
    ax2.hist(G_closeness_cent_non_normalized_val, range = (0,0.02), bins = 100)
    ax2.title.set_text('Histogram of Non-Normalized closseness centrality for '+ name)
    plt.show()
```

### Γ3. Ενδιαμεσική κεντρικότητα (Betweenness centrality)
##### Γ3.1 Να αναπαρασταθεί η κατατανομή της ενδιαμεσικής κεντρικότητας κόμβων για κάθε τοπολογία.

##### Να σχολιαστούν τα αποτελέσματα.

```python
def find_betweenness(G,name):
    print(name)
    G_betweenness = nx.betweenness_centrality(G)
    G_betweenness_val = np.array(list(G_betweenness.values())) 


    G_betweenness_non_normalized = nx.betweenness_centrality(G, normalized = False)
    G_betweenness_non_normalized_val = np.array(list(G_betweenness_non_normalized.values())) 
    
    print("Average betweenness centrality:", np.mean(G_betweenness_non_normalized_val))
    print("Variance of betweenness centrality:",np.var(G_betweenness_non_normalized_val))
    print("Average normalized betweenness centrality:", np.mean(G_betweenness_val))
    print("Variance of normalized betweenness centrality:",np.var(G_betweenness_val))
    fig = plt.figure(figsize=(12, 6))
    ax1 = fig.add_subplot(121)
    ax2 = fig.add_subplot(122)
    ax1.set_ylabel("# of nodes")
    ax1.set_xlabel("Normalized betweenness centrality")
    ax2.set_ylabel("# of nodes")
    ax2.set_xlabel("betweenness centrality")
    ax1.hist(G_betweenness_val, range = (0,1), bins = 100)
    ax1.title.set_text('Histogram of Normalized betweenness centrality for '+ name)
    ax2.hist(G_betweenness_non_normalized_val, range = (0,(0.2*(n-1)*(n-2)/2)), bins=100)
    ax2.title.set_text('Histogram of Non-Normalized betweenness centrality for '+ name)
    plt.show()
```

### Γ4. Kεντρικότητα Katz
##### Γ4.1 Να αναπαρασταθεί η κατανομή της κεντρικότητας Katz των κόμβων για κάθε τοπολογία. 
Για τον προσδιορισμό της παραμέτρου $alpha$ της συνάρτησης katz_centrality,  να υπολογιστεί η μεγαλύτερη ιδιοτιμή $λ_{max}$ του πίνακα γειτνίασης με τη συνάρτηση adjacency_spectrum του networkx. 
H $alpha$ ορίζεται να είναι ιση με $\frac{1}{λ_{max}}-0.01$.
##### Γ4.2 Να υπολογιστεί η μέση  κεντρικότητα Katz  καθώς και η διασπορά της για κάθε τοπολογία.
##### Να σχολιαστούν τα αποτελέσματα.

```python
def find_katz(G, name):
    print(name)
    a = (1/(nx.adjacency_spectrum(G).max()) - 0.01).real
    G_katz_centrality = nx.katz_centrality(G,a)
    G_katz_centrality_val = np.array(list(G_katz_centrality.values())) 
    print("Average Katz centrality:", np.mean(G_katz_centrality_val))
    print("Variance of Katz centrality:",np.var(G_katz_centrality_val))

    plt.figure(figsize=(12, 6))
    plt.hist(G_katz_centrality_val, bins=100)
    plt.title('Histogram of Katz centrality for ' + name)
    plt.ylabel("# of nodes")
    plt.show()

```

```python
find_katz(REG,"REG")
find_katz(RGER,"RGER")
find_katz(RGG,"RGG")
find_katz(SF,"SF")
find_katz(SW,"SW")
```

Παρατηρήσεις:
Η ενδιάμεση κεντρικότητα Katz (betweenness centrality) προσδιορίζει τη σημαντικότητα κάθε κόμβου με κριτήριο τον αριθμό των όλων των δυνατών διαδρομών στις οποίες συμμετέχει


* Για τον REG γράφο λόγω ομοιομορφίας, όλοι οι κόμβοι έχουν ίδια κεντρικότητα Katz

* Για όλους τους υπόλοιπους γράφους παρατηρείται παρόμοια συμπεριφορά με τους κόμβους να έχουν γενικά διαφορετικές τιμές (σε λίγες περιπτώσεις δύο ή περισσότεροι κόμβοι έχουν την ίδια τιμή) ενώ παράλληλα το εύρος των τιμών παραμένει περίπου ίδιο από 0 μέχρι 0.3

#### Να προσδιοριστούν, για κάθε τοπολογία, οι πιο κεντρικοί κόμβοι ως προς το βαθμό, την εγγύτητα, την ενδιαμεσική κεντρικότητα και την κεντρικότητα Katz. Να σχολιαστούν τα αποτελέσματα.

```python
def find_central(G, name):
    print(name)
    G_deg_cent = nx.degree_centrality(G)
    print("Most central node(s) based on degree centrality",{k for k,v in G_deg_cent.items() if v == max(G_deg_cent.values())})
    G_closeness_cent = nx.closeness_centrality(G)
    print("Most central node(s) based on closeness centrality",{k for k,v in G_closeness_cent.items() if v == max(G_closeness_cent.values())})
    G_betweenness = nx.betweenness_centrality(G)
    print("Most central node(s) based on betweenness centrality",{k for k,v in G_betweenness.items() if v == max(G_betweenness.values())})
    a = (1/(nx.adjacency_spectrum(G).max()) - 0.01).real
    G_katz_centrality = nx.katz_centrality(G,a)
    print("Most central node(s) based on Katz centrality",{k for k,v in G_katz_centrality.items() if v == max(G_katz_centrality.values())})
```

```python
find_central(REG,"REG")
find_central(RGER,"RGER")
find_central(RGG,"RGG")
find_central(SF,"SF")
find_central(SW,"SW")
```

## E) Μελέτη της εξελικτικής μετατροπής δικτύου REG σε δίκτυο SW και RGER

##### Μεταβάλλοντας την παράμετρο $g_p$ από $0$ έως $1$ με βήμα $0.1$της συνάρτησης που κατασκευάζει τοπολογίες SW μεγέθους $n=30x$ ( αντικαταστήστε το x με το τελευταίο ψηφίο του ΑΜ) με βαθμό αρχικού πλέγματος $d=4$, να βρείτε το μέσο μήκος μονοπατιού και το μέσο συντελεστή ομαδοποίησης. Πώς συμπεριφέρεται το παραγόμενο δίκτυο για $g_p = 0$, πώς για $g_p = 1$ και πώς για ενδιάμεσες τιμές $g_p \simeq 0.1 - 0.5$;

```python
random.seed(seedno)
new_n=301
d=4

for gp in range(0,11,1):
    print("gp =",gp/10)
    SW_temp = nx.watts_strogatz_graph(n=new_n,k=d,p=gp/10) 
    
    paths = dict(nx.all_pairs_shortest_path_length(SW_temp))
    paths_flattened = pd.json_normalize(paths, sep='-').to_dict(orient='records')[0]
    paths_flattened_val = np.array(list(paths_flattened.values())) 
    print("Average path length:", np.mean(paths_flattened_val))

    G_clust_coef = nx.clustering(SW_temp)
    G_clust_coef_val = np.array(list(G_clust_coef.values())) 
    print("Mean value of clustering coefficients:", np.mean(G_clust_coef_val))
```

Παρατηρήσεις:
* Για $ g_p = 0 $ ο γράφος είναι REG
* Για $ g_p = 0 $ ο γράφος είναι πλήρως τυχαίος - Random
* Για $g_p \simeq 0.1 - 0.5$ ο γράφος είναι SW
